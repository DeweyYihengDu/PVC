```python
import os
import csv

# 设置文件夹路径和新CSV文件的路径
folder_path = './'  # 替换为你的文件夹路径
output_file = 'output.csv'  # 替换为你的输出文件路径

# 获取文件夹中以结尾的文件列表
file_list = [filename for filename in os.listdir(folder_path) if filename.endswith('.annotations')]

# 定义表头字段
header = ['name','query','seed_ortholog','evalue','score','eggNOG_OGs','max_annot_lvl','COG_category','Description','Preferred_name','GOs','EC','KEGG_ko','KEGG_Pathway','KEGG_Module','KEGG_Reaction','KEGG_rclass','BRITE','KEGG_TC','CAZy','BiGG_Reaction','PFAMs']

# 打开新的CSV文件并创建写入器
with open(output_file, 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)

    # 写入CSV文件的表头
    writer.writerow(header)

    # 遍历文件列表并逐个处理文件
    for filename in file_list:
        file_path = os.path.join(folder_path, filename)
        file_prefix = filename.split('.')[0]
        # 打开文件并按行读取内容
        with open(file_path, 'r') as file:
            content = file.readlines()

        # 将每一行内容写入CSV文件的一行
        for line in content:
            if not line.startswith("#"):
                row = [file_prefix] + line.strip().split('\t')
                # row = line.strip().split('\t')

                # 将处理后的行写入CSV文件
                writer.writerow(row)

print('文件整合完成。')
```
