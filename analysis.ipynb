```python
from Bio.Blast import NCBIWWW
from Bio import SeqIO
from Bio.Blast import NCBIXML

# 读取fasta文件
filename = "16S_sequences.fasta"
fasta_sequences = SeqIO.parse(open(filename), "fasta")

# 对每个序列进行BLAST分析
for seq_record in fasta_sequences:
    print(f"Performing BLAST for {seq_record.id}...")

    # 使用16S rRNA数据库进行BLAST分析
    result_handle = NCBIWWW.qblast("blastn", "nt", seq_record.seq, entrez_query="16S rRNA", hitlist_size=10)

    # 解析BLAST结果
    blast_records = NCBIXML.parse(result_handle)

    for blast_record in blast_records:
        print(f"\nResults for {blast_record.query}:\n")
        for alignment in blast_record.alignments:
            for hsp in alignment.hsps:
                print(f"Match: {alignment.title}")
                print(f"Identity: {hsp.identities}/{hsp.align_length}")
                print(f"Score: {hsp.score}")
                print(f"E-value: {hsp.expect}")

    result_handle.close()

```
