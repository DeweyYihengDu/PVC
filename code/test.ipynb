```python
import requests

def download_large_file(url, file_path):
    with requests.get(url, stream=True) as r:
        r.raise_for_status()
        with open(file_path, 'wb') as f:
            for chunk in r.iter_content(chunk_size=8192): 
                f.write(chunk)

url = 'https://microbeatlas.org/downloads/otutables/samples-otus.99.mapped.gz'
file_path = '../samples-otus.99.mapped.gz' 

download_large_file(url, file_path)
```

```python
large_file_path = '../samples-otus.99.metag.minfilter'

file_count = 0
output_file = None

with open(large_file_path, 'r') as file:
    for line in file:

        if line.startswith('>'):

            if output_file is not None:
                output_file.close()

  
            file_count += 1
            new_file_name = f'subfile_{file_count}.txt'
            

            output_file = open(new_file_name, 'w')

        if output_file is not None:
            output_file.write(line)

if output_file is not None:
    output_file.close()
```
