```python
import os
from Bio import Entrez, SeqIO

def download_genomes(taxon_name, output_file):
    Entrez.email = "u7457260@anu.edu.au"  # 请使用你的邮箱地址
    query = f"{taxon_name}[Organism] AND bacteria[Organism] AND (complete genome[Title] OR chromosome[Title])"
    search_handle = Entrez.esearch(db="nucleotide", term=query, retmax=100000)
    search_results = Entrez.read(search_handle)
    search_handle.close()

    id_list = search_results["IdList"]

    if not id_list:
        print("没有找到相关的基因组序列")
        return

    with open(output_file, "w") as output_handle:
        for genome_id in id_list:
            fetch_handle = Entrez.efetch(db="nucleotide", id=genome_id, rettype="fasta", retmode="text")
            seq_record = SeqIO.read(fetch_handle, "fasta")
            fetch_handle.close()
            SeqIO.write(seq_record, output_handle, "fasta")

if __name__ == "__main__":
    taxon_name = "Chlamydiae"  # 你要查询的门名称
    output_file = "Chlamydiae_genomes.fasta"  # 输出的FASTA文件名
    download_genomes(taxon_name, output_file)
```
